```python
import pandas as pd
```

```python
# to concat lecture-professor-class time
sw_reviews = pd.read_csv('./sw_reviews_with_star.csv')
review_professor = sw_reviews[1:]
review_professor = review_professor.drop(['review', 'clean_text', 'star'], axis=1)
review_professor['lecture'][review_professor['lecture']=='문제해결및실습C++'] = '문제해결및실습:C++'
review_professor = review_professor.drop_duplicates().reset_index().drop(['index'], axis=1)
review_professor
```

```python
all_professor = pd.read_csv('./crawling.csv')
all_professor = all_professor.drop(['Unnamed: 0'], axis=1)
all_professor = all_professor[['과목명', '교수명']]
all_professor.columns = ['lecture', 'professor']

all_professor
```

```python
no_review_professor = pd.concat([review_professor, all_professor]).drop_duplicates(keep=False)
no_review_professor
```

```python
no_review_professor.to_csv('no_review_professor.csv')
```

```python
crawling = pd.read_csv('./crawling.csv').drop(['Unnamed: 0'], axis=1)
crawling.columns = ['lecture', 'professor', 'class time']

crawling
```

```python
sw_grouped = pd.read_csv('./sw_review_grouped_without_class_time.csv').drop(['Unnamed: 0'], axis=1)
sw_grouped
```

```python
lecture_with_classtime = pd.merge(left = crawling , right = sw_grouped, how='left', on=['lecture', 'professor'])
lecture_with_classtime = lecture_with_classtime.sort_values(['lecture', 'professor'], ascending=[True, True])
display(lecture_with_classtime)

lecture_with_classtime.to_csv('lecture_with_classtime.csv')
```
