```python
train = pd.read_csv('./processed_lec_review.csv')
test = pd.read_csv('./test_lec_review.csv')
```

```python
display(train.head())
display(test.head())
```

```python
X_train = train[['clean_review', 'pos_neg']]
X_train.head()
```

```python
y_train = train[['pos_neg']]
y_train
```

```python
X_test = test[['clean_review']]
X_test.head()
```

```python
X_train_text = list(X_train['clean_review'])
X_test_text  = list(X_test['clean_review'])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=10000)
train_term = tfidf.fit_transform(X_train_text)
test_term = tfidf.transform(X_test_text)
```

```python
from sklearn.svm import LinearSVC

svm = LinearSVC(C= 1.0, penalty = "l2", random_state= 42)
svm.fit(train_term, y_train.values.ravel())

pred = svm.predict(test_term)
```

```python
df = pd.DataFrame({'review': X_test_text, 'pos_neg': pred})
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

cv = CountVectorizer(lowercase=True, max_df=1.0)
train_term = cv.fit_transform(X_train_text)
test_term = cv.transform(X_test_text)

nb = MultinomialNB(alpha = 0.25, fit_prior=False)
nb.fit(train_term, y_train.values.ravel())

pred = nb.predict(test_term)
```

```python
df = pd.DataFrame({'review': X_test_text, 'pos_neg': pred})
df.head()
```
